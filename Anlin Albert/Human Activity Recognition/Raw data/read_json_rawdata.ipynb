```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/anlinalbert/Main-Project/master/Anlin%20Albert/Raw%20data/Walking%20-%20Example.json'
dataset = pd.read_json(url)
```

```python
dataset.head(10)
```

```python
values = dataset.payload.values[-1]
df = pd.DataFrame(values)
accx = df.iloc[:,0]
accy = df.iloc[:,1]
accz = df.iloc[:,2]
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.plot(accx, label='X-axis')
plt.plot(accy, label='Y-axis')
plt.plot(accz, label='Z-axis')
plt.legend()
```
